```python
import lasio as las
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
import missingno as msno


#step 1: describing the data

poro = pd.read_csv('poro_perm_data.csv')



msno.matrix(poro)

missing_values = poro.isnull().sum()
```

```python
print(missing_values)
```

```python
poro.dropna(inplace=True)
```

```python
poro.isnull().sum()
```

```python
msno.matrix(poro)
```

```python
poro.describe()
```

```python
drop1 = poro[(poro['Porosity (%)' ] < 0)].index
poro.drop(drop1, inplace=True)

drop2 = poro[(poro['Porosity (%)' ] > 100)].index
poro.drop(drop2, inplace=True)

drop3 = poro[(poro['Permeability (mD)' ] <= 0)].index
poro.drop(drop3, inplace=True)

drop4 = poro[(poro['Depth (ft)' ] < 0)].index
poro.drop(drop4, inplace=True)
```
